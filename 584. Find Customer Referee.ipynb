### 584. Find Customer Referee

## Find the names of the customer that are either:

##referred by any customer with id != 2.
##not referred by any customer.

```python
%sql
CREATE TABLE Customer584 (
    id INT PRIMARY KEY,
    name VARCHAR(50),
    referee_id INT 
);
```

```python
%sql
INSERT INTO Customer584 (id, name, referee_id)
VALUES 
    (1, 'Will', NULL),
    (2, 'Jane', NULL),
    (3, 'Alex', 2),
    (4, 'Bill', NULL),
    (5, 'Zack', 1),
    (6, 'Mark', 2);
```

```python
%sql
SELECT name
FROM Customer584
WHERE referee_id != 2 
   OR referee_id IS NULL;
```
