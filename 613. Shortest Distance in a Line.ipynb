### 613. Shortest Distance in a Line 

### Find the shortest distance between any two points from the Point table.

#The result format is in the following example.

```python
%sql
CREATE TABLE Point (
    x INT 
);
```

```python
%sql
INSERT INTO Point (x)
VALUES 
    (-1),
    (0),
    (2);
```

```python
%sql
select * from Point;
```

```python
%sql
select 
min(abs(p2.x-p.x)) as shortest
from Point p join point p2 on p.x > p2.x;
    
```
