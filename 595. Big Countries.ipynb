```python
%sql
drop table world;
```

```python
%sql
CREATE TABLE World (
    name VARCHAR(255) ,
    continent VARCHAR(255),
    area INT,
    population BIGINT,
    gdp BIGINT
);
```

### A country is big if:

#it has an area of at least three million (i.e., 3000000 km2), or
#it has a population of at least twenty-five million (i.e., 25000000).
