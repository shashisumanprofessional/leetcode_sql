### 610. Triangle Judgement

Report for every three line segments whether they can form a triangle.

Return the result table in any order.

```python
%sql
CREATE TABLE Triangle (
    x INT,
    y INT,
    z INT
);
```

```python
%sql
INSERT INTO Triangle (x, y, z)
VALUES 
    (13, 15, 30),
    (10, 20, 15);
```

```python
%sql
select * from triangle;
```
