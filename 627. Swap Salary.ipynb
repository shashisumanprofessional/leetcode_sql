#627. Swap Salary

#Write an SQL query to swap all 'f' and 'm' values (i.e., change all 'f' values to 'm' and vice versa) with a single update statement and no intermediate temp table(s).

```python
%sql
CREATE TABLE Salary (
    id INT,
    name VARCHAR(50),
    sex CHAR(1),
    salary INT
);

INSERT INTO Salary (id, name, sex, salary)
VALUES 
(1, 'A', 'm', 2500),
(2, 'B', 'f', 1500),
(3, 'C', 'm', 5500),
(4, 'D', 'f', 500);
```

```python
%sql
select * from salary;
```

```python
%sql
UPDATE salary
SET sex = (CASE
    WHEN sex = 'm' THEN 'f'
    WHEN sex = 'f' THEN 'm'
END);
```
