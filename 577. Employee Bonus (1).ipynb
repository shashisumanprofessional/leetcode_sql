### 577. Employee Bonus

### Write a solution to report the name and bonus amount of each employee who satisfies either of the following:

##The employee has a bonus less than 1000.
##The employee did not get any bonus.
##Return the result table in any order.

```python
%sql
CREATE TABLE Employee577 (
    empId INT,
    name VARCHAR(255),
    supervisor INT,
    salary INT
);

CREATE TABLE Bonus577 (
    empId INT,
    bonus INT
);
```

```python
%sql
-- Inserting into Employee
INSERT INTO Employee577 (empId, name, supervisor, salary) VALUES (3, 'Brad', NULL, 4000);
INSERT INTO Employee577 (empId, name, supervisor, salary) VALUES (1, 'John', 3, 1000);
INSERT INTO Employee577 (empId, name, supervisor, salary) VALUES (2, 'Dan', 3, 2000);
INSERT INTO Employee577 (empId, name, supervisor, salary) VALUES (4, 'Thomas', 3, 4000);

-- Inserting into Bonus
INSERT INTO Bonus577 (empId, bonus) VALUES (2, 500);
INSERT INTO Bonus577 (empId, bonus) VALUES (4, 2000);
```

```python
%sql
select * from employee577 e left join bonus577 b on e.empId=b.empId  where bonus<1000 or  bonus is null;
```
