### 603. Consecutive Available Seats

###  all the consecutive available seats order by the seat_id using the following cinema table

```python
%sql
CREATE TABLE Cinema (
    seat_id INT,
    free INT
);
```

```python
%sql
INSERT INTO Cinema (seat_id, free) VALUES (1, 1);
INSERT INTO Cinema (seat_id, free) VALUES (2, 0);
INSERT INTO Cinema (seat_id, free) VALUES (3, 1);
INSERT INTO Cinema (seat_id, free) VALUES (4, 1);
INSERT INTO Cinema (seat_id, free) VALUES (5, 1);
```

```python
%sql
select * from Cinema order by seat_id;
```

```python
%sql
select distinct c1.seat_id from cinema c1 join Cinema c2 on abs(c1.seat_id - c2.seat_id) =1 where  c1.free = 1 and c2.free=1;
    
--select * from _sqldf order by seat_id;
```
